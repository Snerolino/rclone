```python
import os
import pandas as pd
from google.colab import drive

# 1. Montar o Google Drive (opcional, mas recomendado para salvar arquivos)
# drive.mount('/content/drive')
# Você pode definir um caminho para o seu projeto no Drive ou usar o ambiente temporário do Colab.
# project_root = "/content/drive/MyDrive/projeto_lora_samara_v1"
project_root = "/content/projeto_lora_samara_v1" # Usando o ambiente temporário do Colab

# Criar o diretório raiz do projeto
os.makedirs(project_root, exist_ok=True)

# Definir os caminhos para as subpastas
images_dir = os.path.join(project_root, "images")
captions_dir = os.path.join(project_root, "captions")
os.makedirs(images_dir, exist_ok=True)
os.makedirs(captions_dir, exist_ok=True)

csv_detailed_path = os.path.join(project_root, "captions_detailed.csv")
csv_tags_path = os.path.join(project_root, "captions_tags.csv")


# --- 2. Geração das Legendas Detalhadas (Reaproveitando sua lógica original) ---
bible = (
    "young black woman, dark brown skin with warm undertones, natural textured afro hair, "
    "slim build, photorealistic, ultra-detailed skin texture, pores visible, confident subtle expression"
)

entries = [
    {"filename": "001_headshot_beach_01.jpg", "caption": f"{bible}, headshot, facing camera, slight head tilt, soft natural backlight, beach background, shallow depth of field, shot on Sony A1 85mm f/1.4, cinematic rim light"},
    {"filename": "002_closeup_beach_02.jpg", "caption": f"{bible}, close-up portrait, direct gaze, wet hair strands, golden hour sunlight, visible pores, ultra detailed, 50mm lens"},
    {"filename": "003_threequarter_stool_studio_01.jpg", "caption": f"{bible}, three-quarter shot, seated on stool, relaxed posture, studio softbox key light, subtle rim light, high fidelity, 70mm f/2.0"},
    {"filename": "004_halfbody_street_01.jpg", "caption": f"{bible}, medium shot (waist up), casual streetwear, urban background blurred, candid pose looking away, overcast soft light, 35mm lens, documentary style"},
    {"filename": "005_fullbody_boardwalk_01.jpg", "caption": f"{bible}, full body, standing on wooden boardwalk, confident stance, wind moving hair, natural light, environmental portrait, 24mm lens, cinematic color grading"},
    {"filename": "006_sitting_crosslegs_beach_01.jpg", "caption": f"{bible}, seated cross-legged (ground), relaxed pose, beach shore behind, soft diffused daylight, shallow DOF, 85mm lens, editorial mood"},
    {"filename": "007_profile_headshot_studio_01.jpg", "caption": f"{bible}, profile headshot (side view), serene expression, controlled studio rim light, high detail on hair texture, 85mm f/1.4"},
    {"filename": "008_smile_candid_cafe_01.jpg", "caption": f"{bible}, candid smile, indoor cafe environment, warm ambient light, bokeh background, 50mm lens, lifestyle photography"},
    {"filename": "009_leaning_wall_urban_01.jpg", "caption": f"{bible}, three-quarter shot leaning against wall, gritty urban texture behind, dramatic side lighting, high contrast, 35mm lens, editorial street style"},
    {"filename": "010_looking_over_shoulder_beach_01.jpg", "caption": f"{bible}, head and shoulders, looking over shoulder toward camera, soft backlight and rim, shallow DOF, natural seaside tones, 85mm f/1.4"},
    {"filename": "011_mood_shot_studio_lowkey_01.jpg", "caption": f"{bible}, low-key studio portrait, moody shadows, volumetric light beam, dramatic contrast, ultra detailed skin specular highlights, 70mm lens"},
    {"filename": "012_smoky_backlight_urban_01.jpg", "caption": f"{bible}, medium shot with rim/backlight, slight atmospheric haze, urban alley background, cinematic color grading, 50mm lens, photorealistic"},
    {"filename": "013_laughing_candid_park_01.jpg", "caption": f"{bible}, candid laughing expression, park setting with trees, dappled sunlight, natural candid style, 85mm lens, high fidelity"},
    {"filename": "014_neutral_studio_clean_01.jpg", "caption": f"{bible}, neutral expression, clean minimalist studio backdrop, soft even lighting, full-frame clarity, 85mm f/1.4, editorial beauty"},
    {"filename": "015_hands_near_face_closeup_01.jpg", "caption": f"{bible}, close-up with hands near face, delicate gesture, fine skin texture visible, gentle softbox lighting, 50mm macro-like detail"},
    {"filename": "016_urban_nightlights_portrait_01.jpg", "caption": f"{bible}, portrait at dusk with city lights bokeh, colorful rim highlights, shallow DOF, cinematic neon reflections, 50mm lens, photorealistic"},
    {"filename": "017_fullbody_sitting_bench_01.jpg", "caption": f"{bible}, full body seated on bench, relaxed legs posture, street background softly blurred, natural ambient light, 35mm lens, lifestyle portrait"},
    {"filename": "018_windy_hair_action_beach_01.jpg", "caption": f"{bible}, action moment hair blown by wind, slight motion in hair, sharp face focus, golden hour sunlight, 85mm f/1.4, editorial moment"},
    {"filename": "019_soft_portrait_window_light_01.jpg", "caption": f"{bible}, soft portrait lit by window light, intimate mood, high detail skin and eye catchlights, 50mm lens, documentary beauty"},
    {"filename": "020_highcontrast_shadows_urban_01.jpg", "caption": f"{bible}, dramatic high-contrast portrait, strong side shadows, textured wall background, cinematic tone, 70mm lens"},
    {"filename": "021_overhead_angle_headshot_01.jpg", "caption": f"{bible}, slight overhead angle headshot, contemplative expression, soft fill light, visible hair texture, 35mm lens"},
    {"filename": "022_pensive_look_medium_beach_01.jpg", "caption": f"{bible}, medium shot with pensive look, muted color palette, shoreline blurred behind, soft cloudy light, 85mm lens"},
    {"filename": "023_adorable_halfsmile_studio_01.jpg", "caption": f"{bible}, half-smile portrait, controlled rim and key light, beauty retouch minimal, ultra-detailed pores and eyelashes, 85mm f/1.4"},
    {"filename": "024_casual_walk_fullbody_street_01.jpg", "caption": f"{bible}, full body walking shot, motion mid-step, urban avenue background, natural ambient light, 35mm lens, candid editorial"},
    {"filename": "025_eyes_closed_serene_beach_01.jpg", "caption": f"{bible}, headshot with eyes closed, serene expression, soft warm backlight, detailed skin texture, 50mm lens, photorealistic"},
    {"filename": "026_portrait_blackwhite_highdetail_01.jpg", "caption": f"{bible}, black and white high-detail portrait, strong texture emphasis, dramatic lighting, film-like grain subtle, 85mm lens"},
    {"filename": "027_angled_threequarter_studio_01.jpg", "caption": f"{bible}, three-quarter studio portrait with angled camera, confident posture, soft fill and hair rim light, 70mm lens"},
    {"filename": "028_smoky_sunset_silhouette_01.jpg", "caption": f"{bible}, silhouette-ish rim-lit portrait at sunset, warm glow edge lighting, background sea horizon, shallow DOF, cinematic vibe"},
    {"filename": "029_closeup_eyes_focus_01.jpg", "caption": f"{bible}, intense close-up focusing on eyes, ultra-detailed irises and lashes, minimal makeup, studio catchlights, 100mm macro-like detail"},
    {"filename": "030_environmental_portrait_market_01.jpg", "caption": f"{bible}, environmental portrait in open market, colorful out-of-focus stalls behind, natural directional light, 50mm lens, lifestyle documentary"}
]

df_detailed_captions = pd.DataFrame(entries)
df_detailed_captions.to_csv(csv_detailed_path, index=False, encoding='utf-8')
print(f"CSV de legendas detalhadas salvo em: {csv_detailed_path}")
print("\nPrimeiras 5 legendas detalhadas:")
print(df_detailed_captions.head())

# --- 3. Geração das Legendas Curtas (Tags) ---
def generate_tags(caption):
    # Remover a "bíblia" inicial para focar nas variações
    base_bible = (
        "young black woman, dark brown skin with warm undertones, natural textured afro hair, "
        "slim build, photorealistic, ultra-detailed skin texture, pores visible, confident subtle expression"
    )
    caption_stripped = caption.replace(base_bible + ", ", "").replace(base_bible, "").strip()

    tags = []
    if "headshot" in caption_stripped: tags.append("headshot")
    if "close-up" in caption_stripped: tags.append("close-up")
    if "three-quarter shot" in caption_stripped: tags.append("three-quarter shot")
    if "medium shot" in caption_stripped: tags.append("medium shot")
    if "full body" in caption_stripped: tags.append("full body")
    if "seated" in caption_stripped or "sitting" in caption_stripped: tags.append("sitting")
    if "standing" in caption_stripped: tags.append("standing")
    if "profile" in caption_stripped: tags.append("profile view")
    if "candid" in caption_stripped: tags.append("candid")
    if "leaning against wall" in caption_stripped: tags.append("leaning against wall")
    if "looking over shoulder" in caption_stripped: tags.append("looking over shoulder")
    if "low-key studio" in caption_stripped or "moody shadows" in caption_stripped or "dramatic contrast" in caption_stripped: tags.append("moody lighting")
    if "golden hour" in caption_stripped: tags.append("golden hour")
    if "beach" in caption_stripped: tags.append("beach")
    if "studio" in caption_stripped: tags.append("studio")
    if "urban" in caption_stripped or "street" in caption_stripped or "cafe" in caption_stripped or "market" in caption_stripped: tags.append("urban street scene")
    if "park" in caption_stripped: tags.append("park")
    if "smile" in caption_stripped or "laughing" in caption_stripped: tags.append("smiling")
    if "pensive" in caption_stripped or "contemplative" in caption_stripped: tags.append("pensive")
    if "eyes closed" in caption_stripped: tags.append("eyes closed")
    if "black and white" in caption_stripped: tags.append("black and white")
    if "high-contrast" in caption_stripped: tags.append("high contrast")
    if "cinematic" in caption_stripped: tags.append("cinematic")
    if "photorealistic" in caption_stripped: tags.append("photorealistic")
    if "natural light" in caption_stripped or "soft diffused daylight" in caption_stripped or "overcast soft light" in caption_stripped: tags.append("natural lighting")
    if "window light" in caption_stripped: tags.append("window light")

    # Adicionar a tag de identificação da pessoa
    tags.insert(0, "samara_v1") # Adicionando a tag identificadora no início

    # Limpar tags duplicadas e juntar
    return ", ".join(sorted(list(set(tags))))

df_detailed_captions['tags'] = df_detailed_captions['caption'].apply(generate_tags)
df_tags_only = df_detailed_captions[['filename', 'tags']]
df_tags_only.to_csv(csv_tags_path, index=False, encoding='utf-8')
print(f"CSV de tags curtas salvo em: {csv_tags_path}")
print("\nPrimeiras 5 legendas curtas (tags):")
print(df_tags_only.head())

# --- 4. Criação dos arquivos .txt de legendas na pasta 'captions' ---
print(f"\nGerando arquivos .txt de legendas na pasta '{captions_dir}'...")
for index, row in df_detailed_captions.iterrows():
    filename_base = os.path.splitext(row['filename'])[0]
    caption_filepath = os.path.join(captions_dir, f"{filename_base}.txt")

    # Usamos a legenda DETALHADA para os arquivos .txt que o LoRA geralmente consome
    with open(caption_filepath, "w", encoding='utf-8') as f:
        f.write(row['caption'])

print(f"Total de {len(df_detailed_captions)} arquivos .txt de legendas criados.")
print("\nEstrutura de pastas e arquivos criada:")
!ls -R {project_root}

# --- 5. Simulação de upload de imagens ---
# No ambiente real, você colocaria suas 30 imagens JPEG aqui.
# Para esta simulação, vamos criar 30 arquivos dummy para preencher a pasta.
from PIL import Image
import numpy as np

print(f"\nCriando arquivos de imagem dummy na pasta '{images_dir}'...")
for filename in df_detailed_captions['filename']:
    dummy_image_path = os.path.join(images_dir, filename)
    # Criar uma imagem branca simples como placeholder
    img = Image.fromarray(np.full((512, 512, 3), 255, dtype=np.uint8))
    img.save(dummy_image_path)
print(f"Total de {len(df_detailed_captions)} arquivos de imagem dummy criados.")

print("\nVerificação final da estrutura de pastas:")
!ls -R {project_root}
```
